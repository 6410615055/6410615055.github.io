# Dowload ข้อมูล Data จาก kagglehub เพื่อนำมาใช้งาน

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("iamtanmayshukla/lung-cancer-data")

print("Path to dataset files:", path)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

df = pd.read_csv('/root/.cache/kagglehub/datasets/iamtanmayshukla/lung-cancer-data/versions/1/lung cancer data.csv') # อ่านข้อมูลที่เรามี

df.head() # แสดงข้อมูล 5 แถวแรก
```

```python
df['GENDER'] = encoder.fit_transform(df['GENDER'])
df['LUNG_CANCER'] = encoder.fit_transform(df['LUNG_CANCER'])
df.info() # แสดงชื่อของหัวข้อ จำนวนข้อมูล หน่วยที่มีทั้งหมด
```

```python
df.head(309)
```

```python
# เป้าหมายคืออาการป่วยมะเร็งปลอด
X = df.drop('LUNG_CANCER', axis=1)  # ลบคอลัมน์ target ออก
Y = df['LUNG_CANCER']  # เป้าหมาย (target)

# แบ่งข้อมูลเป็นชุดฝึก (train) และชุดทดสอบ (test) โดย 30% เป็น test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)

#แสดงขนาดของชุดข้อมูลที่เราได้ทำการแบ่ง
print(f"Training data shape: {X_train.shape}")
print(f"Test data shape: {X_test.shape}")
```

# ขั้นตอนต่อไปคือการเลือกและฝึกโมเดลที่คุณต้องการใช้ เช่น
# 1.Logistic Regression
# 2.Linear Regression
# 3.Decision Tree
# 4.Random Forest
# 5.Support Vector Machine (SVM)
# โดยใช้ข้อมูลจาก X_train และ y_train สำหรับการฝึกโมเดล

# **Example : Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression

# สร้างโมเดล
model = LogisticRegression()

# ฝึกโมเดลด้วยข้อมูลชุดฝึก
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

# ทำนายผลลัพธ์จากข้อมูลชุดทดสอบ
Y_pred = model.predict(X_test)

# คำนวณความแม่นยำของโมเดล
accuracy = accuracy_score(Y_test, Y_pred)
print(f"Model Accuracy: {accuracy * 100:.2f}%")
```

# **Example : Linear Regression**

```python
from sklearn.preprocessing import StandardScaler

# สเกลข้อมูล
scaler = StandardScaler()

# สเกล X_train และ X_test
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

# สร้างโมเดล Linear Regression
model = LinearRegression()

# ฝึกโมเดล
model.fit(X_train_scaled, Y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# ทำนายค่าด้วย test set
y_pred = model.predict(X_test_scaled)

# ประเมินโมเดล
mse = mean_squared_error(Y_test, Y_pred)
r2 = r2_score(Y_test, Y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared (R²): {r2}')
```
